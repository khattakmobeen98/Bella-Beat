# DOING ANALYSIS

**Categorizing Activity Levels Based on Total Steps**  

To better understand user activity levels, we can categorize them based on their average daily step counts:  

1. **Sedentary**:  
   - **Average Steps**: Less than 6,000  
   
2. **Lightly Active**:  
   - **Average Steps**: Between 6,000 and 9,000
      
3. **Active**:  
   - **Average Steps**: Between 9,000 and 12,000
     
4. **Very Active**:  
   - **Average Steps**: More than 12,000  



```python
#Grouping the activity by the id and finding the mean of the total steps
id_group=activity.groupby('id')['totalsteps'].mean().sort_values(ascending=False)
#Renamming the columns
id_group.columns=['id','totalsteps']
#Resetting the index
id_group=id_group.reset_index()

#Condition of activity level
condition=[
    (id_group['totalsteps']<6000),
    (id_group['totalsteps']>=6000) & (id_group['totalsteps']<9000),
    (id_group['totalsteps']>=9000) & (id_group['totalsteps']<12000),
    id_group['totalsteps']>=12000
]
#Activity levels
values=['Sedentary','Lightly Active','Active','Very Active']

id_group['activity_level']=np.select(condition, values, default='Unknown')
```

```python
id_group.head()
```

```python
#Adding the activity level to main activity dataset

activity['activity_level']=activity['id'].map(id_group.set_index('id')['activity_level'])
```

```python
activity.head()
```

# DATA VISUALIZATION

```python
# Find the number of participants in each activity level
n_people_per_activity_group = id_group.groupby('activity_level')['id'].count()
activity_levels = n_people_per_activity_group.index

# Define colors for each activity level
colors = ['cornflowerblue', 'lightgreen', 'salmon', 'gold']

# Plotting the bar chart
bars = plt.bar(activity_levels, n_people_per_activity_group, color=colors)
plt.xlabel('The Activity Level')
plt.ylabel('The Number of Participants')
plt.title('The Number of Participants in Each Activity Level',fontweight='bold')

for i, level in enumerate(activity_levels):
    count = n_people_per_activity_group[level]
    plt.text(i, count, f'{count}', ha='center', va='bottom')

plt.legend(bars,activity_levels, title='Activity Levels')

plt.show()
```

The bar chart represents the number of participants categorized into four different activity levels.
The majority of participants fall under the "Sedentary" and "Lightly Active", while fewer participants are categorized as "Active" or "Very Active". This suggests that most individuals in the dataset engage in lower levels of physical activity.



 Users are classified based on how frequently they wear the device:  

- **Always**: Worn for 62 days.  
- **Usual**: Worn between 50 and 62 days.  
- **Often**: Worn between 35 and 50 days.  
- **Sometimes**: Worn between 20 and 35 days.  
- **Rarely**: Worn for fewer than 20 days.  

To determine usage, we calculate the total METs per day. A METs value of zero indicates the device was not worn, while a value greater than zero confirms usage. The dataset is then grouped by user ID, counting the number of days the device was worn to assign each user to a category.

```python
#Making a copy of the METs
new_METs=METs.copy()
new_METs.head()
```

```python
#Adding the Date Column
new_METs['Date']=new_METs['ActivityMinute'].dt.date
#Converting the Id to String
new_METs['Id']=new_METs['Id'].astype(str)
#Dropping the ActivityMinute Column
new_METs=new_METs[['Id','Date','METs']]

new_METs['Date']=pd.to_datetime(new_METs['Date'],format='mixed') #formatting the date column
new_METs.head()
```

```python
#Groupping by Id and return the total number of days
Total_days=new_METs.groupby('Id')['Date'].nunique().sort_values(ascending=False)
Total_days=Total_days.reset_index()
Total_days.columns=['Id','Total_days']
Total_days.head()
```

```python
#Setting the condition
conditions=[
    (Total_days['Total_days']==62),
    (Total_days['Total_days']<=62) & (Total_days['Total_days']>50),
    (Total_days['Total_days']<=50) & (Total_days['Total_days']>35),
    (Total_days['Total_days']<=35) & (Total_days['Total_days']>20),
    (Total_days['Total_days']<=20)
]
values=['Always','Usual','Often','Sometimes','Rarely']

Total_days['Type_of_Participant']=np.select(conditions,values,default='Unknown')
Total_days.head()
```

```python
# Calculate the maximum number of NonZeroMETsDays for each type of participants
participants_each_type=Total_days.groupby('Type_of_Participant')['Id'].count()

type_of_participants=participants_each_type.index

# Plot the horizontal bar chart
plt.barh(type_of_participants,participants_each_type, color='cornflowerblue')

# Add title and labels
plt.title('Type of Participants',fontweight='bold')
plt.xlabel('Number of Participants',fontdict={'fontsize': 12})
plt.ylabel('Type of Participants',fontdict={'fontsize': 12})

# Add legend
plt.legend(['Number of Participants'], loc='upper right')

# Add values on the bars
for index, value in enumerate(participants_each_type.values):
    plt.text(value, index, str(value))

# Show the plot
plt.show()
```

The bar chart represents different types of participants based on their frequency of participation.
The majority of participants fall under the "Always" category, indicating a high level of consistent participation. On the other side, very few participants engage infrequently, with only 1  participant classified as "Rarely" participating.

**Relation Between Total Steps and Calories**

```python
# Define colors for each activity level
colors = {'Sedentary': 'salmon', 'Lightly Active': 'lightgreen', 'Active': 'cornflowerblue', 'Very Active': 'gold'}
# Scatter Plot of Total Steps vs Calories Burned with different colors for activity levels
for level in activity['activity_level'].unique():
    subset = activity[activity['activity_level'] == level]
    plt.scatter(subset['totalsteps'], subset['calories'], label=level, color=colors[level], s=10)  # s=10 for small dots

plt.xlabel('Total Steps')
plt.ylabel('Calories')
plt.title('Total Steps per Day vs Calories Burn per Day',fontweight='bold')
plt.legend(title='Activity Level')
plt.show()

correlation = activity['totalsteps'].corr(activity['calories'])
print('Correlation between Total Steps and Calories Burned:', correlation)
```

There is a clear positive correlation between the total steps taken per day and the number of calories burned. In simple terms, taking more steps leads to burning more calories, which aligns with common expectations.

Leveraging this insight, the Bellabeat app could introduce a feature allowing users to set a calorie-burning goal. The app can then calculate the required number of steps to achieve that target, helping users stay motivated and committed to their fitness goals.

**In which days are the participants active?**

```python
#Creating a new column for the date
hourly_intensity['Date']=hourly_intensity['ActivityHour'].dt.date
#Making a copy of Intensity
new_intensity=hourly_intensity.copy()

#Dropping the ActivityHour and AverageIntensity columns from the new intensity
new_intensity.drop(['ActivityHour','AverageIntensity'],axis=1,inplace=True)

#Changing the Date to datetime
new_intensity['Date']=pd.to_datetime(new_intensity['Date'])
#Addinng a Day column to new_intensity
new_intensity['Day']=new_intensity['Date'].dt.day_name()
hourly_intensity.head()
```

**In which time of day the participants are most active?**

```python
intensity = hourly_intensity.copy()
#Changing the format of the Date column
intensity['Date'] = pd.to_datetime(intensity['Date'])
#Adding a new column for the hour
intensity['Hour'] = intensity['ActivityHour'].dt.hour
#Adding a new column for the time
intensity['Time'] = intensity['Hour'].apply(lambda x: f"{x:02d}:00:00")
#Adding a new column for the day
intensity['Day'] = intensity['Date'].dt.day_name()

#Removing the unwanted columns
intensity.drop(['ActivityHour','AverageIntensity','Hour'],axis=1,inplace=True)

```

```python
intensity_by_day = intensity.groupby('Time')['TotalIntensity'].sum().reset_index()

# Plotting the Graph
plt.figure(figsize=(14, 7))  # Increase the figure size
plt.bar(intensity_by_day['Time'], intensity_by_day['TotalIntensity'],color='cornflowerblue')
plt.xlabel('Time',fontdict={'fontsize': 12})
plt.ylabel('Total Intensity',fontdict={'fontsize': 12})
plt.title('Total Intensity by Time of Day', fontweight='bold')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.show()
```
